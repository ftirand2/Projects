```python
#Import librairies
import requests
import pandas as pd
```

```python
# Creation of a list of cities used for the project with deletion of elements that are not cities
villes= ["Mont Saint Michel","St Malo","Bayeux", "Le Havre", "Rouen", "Paris", "Amiens", "Lille", "Strasbourg", "Colmar", "Eguisheim", "Besancon", "Dijon", "Annecy", "Grenoble", "Lyon", "Bormes les Mimosas", "Cassis", "Marseille", "Aix en Provence", "Avignon", "Uzes", "Nimes", "Aigues Mortes", "Saintes Maries de la mer", "Collioure", "Carcassonne", "Foix", "Toulouse", "Montauban", "Biarritz", "Bayonne", "La Rochelle"]
```

```python
# Creation of the dataframe to retrieve the data requested in the project
cols=['ville_id' , 'nom', 'latitude', 'longitude', 'meteo', 'proba_pluie', 'jour_temperature']
dataset = pd.DataFrame(columns=cols)
dataset
```

```python
#Add Openweathermap API request to the code
for i in range(len(villes)):
    
    print("Queries for the requested city: {}".format(villes[i]))
    
# GPS data recovery (longitude and latitude with Nominatim)
    URL1 = 'https://nominatim.openstreetmap.org/search'
    params = {
        "city": villes[i],
        "country": "France",
        "format": "json"
    }
    
    r = requests.get(URL1, params=params)
    res = r.json()
    

    dataset.loc[i,'ville_id'] = i
    dataset.loc[i,'nom'] = villes[i]
    dataset.loc[i, 'latitude'] = res[0]['lat']
    dataset.loc[i, 'longitude'] = res[0]['lon'] 
    
    #Following Openweathermap API documentation, excluded parameters have been defined here as a dictionary
    params = {
        "lat": dataset.loc[i, 'latitude'],
        "lon": dataset.loc[i, 'longitude'],
        "exclude": "current,minutely,hourly",
        "units": "metric",
        "appid": "2f2685efc16594e365e34e00af802902",
    }
    
    r = requests.get('https://api.openweathermap.org/data/2.5/onecall?', params=params)
    res = r.json()
    print(res)
    
    calcul_pluie = 0
    for x in res['daily']:
        if 'rain' in x.keys():
            calcul_pluie += x['pop'] * x['rain']
         
    # Calculation of the average temperature: the information must be retrieved in the "daily" category then "temp" sub-category then "day
    for x in res['daily']:
        #récupération des données sous 7 jours sous forme de liste d'ou pd.Series et utilisation de list comprehension
        temperature = pd.Series(x['temp']['day'])
        moyenne_temperature = temperature.mean()
        
    # Extract the most probable weather: use of the mode () function to display the most probable value: in the daily category we retrieve the 
    #information in the weather section (first part therefore index 0 and under main section)
    for x in res['daily']:
        weath = pd.Series(x['weather'][0]['main'])
        weather = weath.mode()[0]

    dataset.loc[i,'meteo']= weather
    dataset.loc[i,'proba_pluie']= calcul_pluie
    dataset.loc[i,'jour_temperature']= moyenne_temperature
```

```python
#Classification of cities according to the expected rain in a "rank" column
dataset.loc[:, 'rang'] = dataset['proba_pluie'].rank(method='min').astype(int)
```
